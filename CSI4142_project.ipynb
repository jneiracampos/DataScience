## Split the total duration to hours and minutes

```python
# function to extract the hours
def find_hours(x):
  x = x.replace("M", "").replace("P", "").replace("T", "")
  if "D" in x and "H" in x:
    return 24*int(x[0]) + int(x[2:].split("H")[0])
  if "D" in x:
    return 24*int(x[0])
  else:
    return int(x.split("H")[0])

# function to exctract the minutes
def find_minutes(x):
  x = x.replace("P", "").replace("T", "")
  if "M" in x:
    if "H" in x:
      return int(x.replace("M", "").split("H")[1]) 
    elif "D" in x:
      return int(x.replace("M", "").split("D")[1]) 
    else:
      return int(x.replace("M", ""))
  else:
    return 0

# update the dataframe
df["duration_hours"] = df['travelDuration'].apply(find_hours)
df["duration_minutes"] = df['travelDuration'].apply(find_minutes)

df
```

## Split the connecting flights

As you can see above, some rows are connecting flights which include multiple flights delimited by ||. The code below splits the rows into one row per flight, and splits the fare based on the duration of each flight. 

```python
# some rows are connecting flights and have || seperating the flights
# takes about 25 minutes

# df = df.head(1000)

import warnings
warnings.filterwarnings("ignore")

# each row that can contain a ||
rows_to_change = ["segmentsDepartureTimeEpochSeconds","segmentsDepartureTimeRaw","segmentsArrivalTimeEpochSeconds","segmentsArrivalTimeRaw","segmentsArrivalAirportCode","segmentsDepartureAirportCode","segmentsAirlineName","segmentsAirlineCode","segmentsEquipmentDescription","segmentsDurationInSeconds","segmentsDistance","segmentsCabinCode"]

for index, row in df.iterrows():
    if not row["isNonStop"]:
        # create the new row, and split the flight values
        new_row = row.copy()  

        for n in rows_to_change:
            tmp = row[n].split("||")
            
            if len(tmp) > 2: # we do not want to deal with connecting flights with more than 2 flights (outlier)
              df = df.drop(index)
              break

            df.at[index, n] = tmp[0]
            new_row[n] = tmp[1]
                    
        
        else:
          # find the duration of each flight and the ratios
          dur1 = int(df.at[index, "segmentsDurationInSeconds"])
          dur2 = int(new_row["segmentsDurationInSeconds"])

          ratio1 = dur1/(dur1+dur2)
          ratio2 = dur2/(dur1+dur2)

          # determine the new fare based on the ratio
          df.at[index, "totalFare"] = ratio1*row["totalFare"]
          new_row["totalFare"] = ratio2*new_row["totalFare"]

          # determine the new duration based on the ratio
          tmp_time = df.loc[index, "duration_hours"]*60 + df.loc[index,"duration_minutes"]

          df.at[index, "duration_minutes"] = int(ratio1*float(tmp_time) % 60)
          df.at[index, "duration_hours"] = ratio1*float(tmp_time) // 60

          new_row["duration_minutes"] = int(ratio2*float(tmp_time) % 60)
          new_row["duration_hours"] = ratio2*float(tmp_time) // 60

          # add the new row 
          df = df.append(new_row)
```

```python
# run this
# sort and reset indices
df = df.sort_index()
df = df.reset_index(drop=True)
df.head(7)
```

##  Create the Fact table

```python
new_df
fact = new_df[["ID","segmentsArrivalAirportCode", "segmentsDepartureAirportCode", "segmentsEquipmentDescription", "segmentsAirlineCode", "totalFare"]]
fact
```

## Join the Airport Dimension

```python
airport_df_join = airport_df[["index","IATA"]]


fact1 = fact.merge(airport_df_join, left_on='segmentsArrivalAirportCode', right_on='IATA', how ="left").merge(airport_df_join, left_on='segmentsDepartureAirportCode', right_on='IATA', how ="left")

fact1 = fact1.rename(columns={"index_x": "arrival_airport_id", 
                                          "index_y":"departure_airport_id"
                              })

fact1 = fact1.drop(columns=["segmentsArrivalAirportCode","segmentsDepartureAirportCode", "IATA_x", "IATA_y" ])

fact1
```

## Join the Airline Dimension

```python
airline_df_join = airline_df[["airline_id", "IATA"]]

fact2 = fact1.merge(airline_df_join, left_on='segmentsAirlineCode', right_on='IATA', how ="left")
fact2 = fact2.drop(columns=["segmentsAirlineCode","IATA" ])

fact2
```

## Join the Airplane Dimension

```python
airplane_df_join = airplane_df2[["airplane_id", "name"]]

fact3 = fact2.merge(airplane_df_join, left_on='segmentsEquipmentDescription', right_on='name', how ="left")
fact3 = fact3.drop(columns=["segmentsEquipmentDescription","name"])

fact3
```

## Join the Flight and Booking dimensions

```python
# the flight and booking ids correspond to the fact table ids

fact3["book_info_id"] = fact3["ID"]
fact3["flight_info_id"] = fact3["ID"]
fact3 = fact3.drop("ID", axis=1)

fact3
```
